```python
import json
import matplotlib.pyplot as plt
from pymacro.ingredient import Ingredient, load_food_items
from pymacro.meal import Meal, Dish
from pymacro.menu import Menu

food_items = load_food_items("./data/food_items.json")

iogur_desayuno = Dish.from_ingredients(
    name="Iogur Desayuno",
    ingredients=[
        Ingredient(food_items["Iogur Griego Natural"], 160),
        Ingredient(food_items["Anacardo"], 10),
        Ingredient(food_items["Nuez"], 10),
        Ingredient(food_items["Plátano"], 140),
        Ingredient(food_items["Arándano"], 20)
    ]
)
tostada_desayuno = Dish.from_ingredients(
    name="Tostada Desayuno",
    ingredients=[
        Ingredient(food_items["Pan Integral"], 30),
        Ingredient(food_items["Huevo"], 50),
        Ingredient(food_items["Pechuga Pollo Embutido"], 75),
        Ingredient(food_items["Queso Light Loncha"], 50),   
    ]
)
arroz_hervido = Dish.from_ingredients(
    name="Arroz Hervido",
    ingredients=[
        Ingredient(food_items["Arroz Integral"], 250),
    ]
)
contramuslo_plancha = Dish.from_ingredients(
    name="Contramuslo Plancha",
    ingredients=[
        Ingredient(food_items["Contramuslo Pollo"], 250),
    ]
)


desayuno = Meal.from_dishes(
    name="Desayuno",
    dishes=[iogur_desayuno, tostada_desayuno]
)

comida = Meal.from_dishes(
    name="Comida",
    dishes=[arroz_hervido, contramuslo_plancha],
    quantities=[100, None]
)

menu = Menu.from_meals(
    name="Menu 1",
    meals=[desayuno, comida])
menu.plot_macros(collapse_meals=True)
```
