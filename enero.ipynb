```python
from pymacro.ingredient import Ingredient
from pymacro.dish import Dish
from pymacro.meal import Meal
from pymacro.menu import Menu
```

# **Semana 2**

## Platos generales

```python
## DESAYUNOS
platano_y_anacardos = Dish.from_db(
    name="Plátano y Anacardos",
    ing_names=["Plátano", "Anacardo"],
    quantities=[140, 10],
)
batido_kefir = Dish.from_db(
    name="Batido Kéfir",
    ing_names=[
        "Kéfir",
        "Frutos Rojos Congelados",
        "Fresas Congeladas",
        "Nuez",
        "Anacardo",
    ],
    quantities=[200, 50, 50, 10, 10],
)
tostada_jamón = Dish.from_db(
    name="Tostada Jamón",
    ing_names=[
        "Pan Integral",
        "Queso Light Loncha",
        "Huevo",
        "Jamón Serrano",
    ],
    quantities=[60, 50, 110, 40],
)
tostada_pollo = Dish.from_db(
    name="Tostada Pollo Carloteña",
    ing_names=[
        "Pan Integral",
        "Queso Light Loncha",
        "Huevo",
        "Pollo Carloteña Embutido",
    ],
    quantities=[25, 50, 110, 65],
)

## COMIDAS
wok_verduras = Dish.from_db(
    name="Wok Verduras",
    ing_names=["Brócoli", "Cebolla", "Judía Verde", "Zanahoria", "Variado de Setas"],
    quantities=[300, 250, 110, 100, 200],
)
filete_pavo = Dish.from_db(
    name="Filete Pavo", ing_names=["Filete Pavo"], quantities=[300]
)
ensalada_pollo = Dish.from_db(
    name="Ensalada Pollo",
    ing_names=["Carne Picada Pollo", "Quinoa Cocida", "Lentejas Cocidas", "Tomate", "Aguacate", "Cebolla"],
    quantities=[250, 130, 120, 70, 60, 20],
)

## CENAS
crema_verduras = Dish.from_db(
    name="Crema Verduras",
    ing_names=[
        "Calabacín",
        "Cebolla",
        "Patata",
        "Zanahoria",
        "Agua de Coco Light",
        "Agua",
    ],
    quantities=[1500, 200, 300, 180, 400, 200],
)
tortilla = Dish.from_db(
    name="Tortilla",
    ing_names=["Huevo", "Pavo Embutido", "Lomo Embutido"],
    quantities=[160, 50, 30],
)
filete_panga = Dish.from_db(
    name="Filete Panga", ing_names=["Filete Panga"], quantities=[250]
)
brocoli_boniato = Dish.from_db(
    name="Brocoli y boniato",
    ing_names=["Brócoli", "Boniato"],
    quantities=[200, 200]
)

## SNACKS
bebida_avellana = Dish.from_db(
    name="Vaso Bebida Avellana", ing_names=["Bebida Avellana"], quantities=[200]
)


```

## **Lunes**

```python
desayuno = Meal(name="Desayuno", dishes=[platano_y_anacardos, tostada_jamón],)
comida = Meal(name="Comida", dishes=[wok_verduras.extract_portion(200), filete_pavo])
cena = Meal(name="Cena", dishes=[tortilla, crema_verduras.extract_portion(200)])
lunes = Menu(name="Lunes", meals=[desayuno, comida, cena])
lunes.plot_macros(collapse_meals=True)
```

## **Martes**

```python
desayuno = Meal(
    name="Desayuno",
    dishes=[batido_kefir, tostada_pollo],
)
comida = Meal(
    name="Comida",
    dishes=[ensalada_pollo],
)
cena = Meal(name="Cena", dishes=[brocoli_boniato, filete_panga])
entrehoras = Meal(name="Entrehoras", dishes=[bebida_avellana, bebida_avellana])
lunes = Menu(name="Lunes", meals=[desayuno, comida, cena, entrehoras])
lunes.plot_macros(collapse_meals=True)
```
